```python
import os

def count_images_in_directory(directory):
    subdir_count = 0
    total_image_count = 0
    images_count_per_subdir = {}

    for subdir in os.listdir(directory):
        subdir_path = os.path.join(directory, subdir)
        if os.path.isdir(subdir_path):
            subdir_count += 1
            image_count = sum([len(files) for r, d, files in os.walk(subdir_path) if files])
            images_count_per_subdir[subdir] = image_count
            total_image_count += image_count

    return subdir_count, images_count_per_subdir, total_image_count

# Specify the path to the 'train' directory
name="train"
directory = '/content/drive/MyDrive/plant_disease_detection/dataset/'+name

# Get the counts
subdir_count, images_count_per_subdir, total_image_count = count_images_in_directory(directory)
start = "\033[1m"
end = "\033[0;0m"
# Print the results
print("Before preprocessing: "+ name+ " set\n")
print(f"Number of subdirectories: {subdir_count}\n")
print(f"Number of images in each subdirectory\n")
for subdir, count in images_count_per_subdir.items():
    print(f"{subdir}: {start}{count}{end}")
print(f"Total number of images in 'train' directory: {total_image_count}")
```

```python
import os
from tabulate import tabulate

def count_images_in_directory(directory):
    subdir_count = 0
    total_image_count = 0
    images_count_per_subdir = {}

    for subdir in os.listdir(directory):
        subdir_path = os.path.join(directory, subdir)
        if os.path.isdir(subdir_path):
            subdir_count += 1
            image_count = sum([len(files) for r, d, files in os.walk(subdir_path) if files])
            images_count_per_subdir[subdir] = image_count
            total_image_count += image_count

    return subdir_count, images_count_per_subdir, total_image_count

# Specify the path to the 'train' directory
name = "train"
directory = '/content/drive/MyDrive/plant_disease_detection/dataset/' + name

# Get the counts
subdir_count, images_count_per_subdir, total_image_count = count_images_in_directory(directory)
start = "\033[1m"
end = "\033[0;0m"
# Print the results
print("\nAfter preprocessing: " + name + " set\n")
print(f"Number of subdirectories: {subdir_count}\n")

# Prepare data for the table
table_data = []
for subdir, count in images_count_per_subdir.items():
    table_data.append([subdir, f"{start}{count}{end}"])

# Print the table
print("Number of images in each subdirectory\n")
print(tabulate(table_data, headers=["Subdirectory", "Image Count"], tablefmt="grid"))

# Print total number of images
print(f"\nTotal number of images in {name} directory: {start}{total_image_count}{end}")
```
