```python
import cv2
import numpy as np
import shutil
import os

def is_black_corner(image, region_size=20):
    h, w = image.shape[:2]
    corners = [
        image[0:region_size, 0:region_size],  # top-left
        image[0:region_size, w-region_size:w],  # top-right
        image[h-region_size:h, 0:region_size],  # bottom-left
        image[h-region_size:h, w-region_size:w]  # bottom-right
    ]
    black = np.array([0, 0, 0])
    black_count = 0

    for corner in corners:
        if np.all(corner == black):
            black_count += 1

    return black_count > 1

# Define paths
path='train'
train_dir = '/content/drive/MyDrive/plant_disease_detection/dataset/' + path  # Update this with your train directory path
invalid_images_dir = '/content/drive/MyDrive/plant_disease_detection/dataset/invalid_images/' + path  # Update this with your invalid images directory path

# Create invalid_images directory if it doesn't exist
os.makedirs(invalid_images_dir, exist_ok=True)

# Iterate over each class folder in the train directory
for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    if os.path.isdir(class_path):
        # Create corresponding class folder in invalid_images directory
        invalid_class_path = os.path.join(invalid_images_dir, class_name)
        os.makedirs(invalid_class_path, exist_ok=True)

        # Iterate over each image in the class folder
        for file_name in os.listdir(class_path):
            if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                file_path = os.path.join(class_path, file_name)
                image = cv2.imread(file_path)
                if is_black_corner(image):
                    # Move image to the invalid_class_path
                    shutil.move(file_path, os.path.join(invalid_class_path, file_name))

print("Processing complete.")
```
